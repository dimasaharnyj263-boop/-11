В) Дано целое число, состоящее из разных цифр. Определить наибольшую цифру числа и вывести её.

```python
# Задание A: таблица умножения
def mult_table(n):
    for i in range(1,11):
        print(f"{n} x {i} = {n*i}")

mult_table(15)
```

```python
# Задание B: найти наибольшую цифру числа
def max_digit(n):
    n = abs(int(n))
    print('Число:', n)
    print('Наибольшая цифра:', max(int(d) for d in str(n)))

max_digit(5731)
```

Ответы на кр вопросы:

1.Общий синтаксис цикла с параметром Цикл с параметром в Python используется для выполнения действий заданное количество раз. Он основан на конструкции for, которая перебирает элементы последовательности (списка, строки, диапазона и т. д.). Общий синтаксис: for переменная in последовательность: Внутри цикла можно обращаться к текущему элементу последовательности. Пример: for i in range(5): print(i) — цикл выполняется 5 раз.

2.Функция range() Функция range() создаёт последовательность чисел, которую часто используют в циклах for. Она может принимать один, два или три аргумента: начало, конец и шаг. Если указан один аргумент, счёт идёт от 0 до этого числа (не включая его). Например, range(3) создаёт числа 0, 1 и 2. range(1, 5, 2) создаёт 1, 3 — то есть шаг равен 2.

3.Функция enumerate() Функция enumerate() используется, когда нужно получать не только элементы последовательности, но и их порядковые номера. Она возвращает пары вида (индекс, значение). Например, при переборе списка ['a','b','c'] через enumerate() можно получить (0,'a'), (1,'b'), (2,'c'). Это удобно для нумерации элементов прямо в цикле.


4.Общий синтаксис цикла с условием Цикл с условием выполняется, пока условие остаётся истинным. Используется конструкция while. Проверка условия происходит перед каждой итерацией, и если оно ложно — цикл завершается. Пример: переменная увеличивается на 1, пока не достигнет нужного значения. Такой тип цикла подходит, когда количество повторений заранее неизвестно.
