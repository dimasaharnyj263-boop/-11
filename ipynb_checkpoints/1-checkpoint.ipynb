А) 2x + 3x / (4 + 0,3*7y)  x=2; y=3  (в тетради выражение частично нечёткое; здесь использована разумная интерпретация)

В) Классификация зданий по высоте:
- до 30 м - здания повышенной этажности;
- до 50 м - I категория;
- до 75 м - II категория;
- до 100 м - III категория;
- более 100 м - высотные здания.

```python
# Задание A: вычислим предложенное выражение при x=2,y=3
x=2
y=3
res = (2*x + 3*x) / (4 + 0.3*7*y)
print('Результат выражения:', res)
```

```python
# Задание B: функция классификации зданий по высоте
def classify_building(height):
    if height <= 30:
        return 'здания повышенной этажности'
    elif height <= 50:
        return 'I категория'
    elif height <= 75:
        return 'II категория'
    elif height <= 100:
        return 'III категория'
    else:
        return 'высотные здания'

# Пример использования:
for h in [25,45,70,95,150]:
    print(h, 'м ->', classify_building(h))
```
