## Варианты заданий — Вариант 9 (реализация)

А) В матрице поменять местами наибольшие элементы в первом и последнем столбцах.

В) Вычислить коэффициент корреляции между элементами первого и последнего столбца.

```python
import numpy as np
np.random.seed(0)
A = np.random.randint(-10,50,(6,6))
print('Матрица A:\n',A)
# swap
i0 = np.argmax(A[:,0])
inlast = np.argmax(A[:,-1])
A[i0,0], A[inlast,-1] = A[inlast,-1], A[i0,0]
print('\nA после обмена:\n',A)
print('\nКоэффициент корреляции между первым и последним столбцами:')
print(np.corrcoef(A[:,0], A[:,-1])[0,1])
```

Ответы на вопросы:

1.Библиотека NumPy. Импорт библиотеки NumPy — это библиотека Python для работы с массивами и научными вычислениями. Она обеспечивает высокую скорость и удобство при работе с большими наборами числовых данных. Импорт выполняется так: import numpy as np



2.Массивы. Структура. Примеры Массив в NumPy — это упорядоченная структура данных, содержащая элементы одного типа. В отличие от списков Python, массивы позволяют выполнять математические операции сразу со всеми элементами. Пример: np.array([1, 2, 3])

3.Создание массива. Функции array() и values() Функция array() создаёт массив из списка или кортежа. Синтаксис: np.array(последовательность) Пример: np.array([1, 2, 3, 4]) Функция values() используется при работе с данными из таблиц (например, из pandas) для получения массива значений.

4.Функции создания массива заданного вида NumPy имеет встроенные функции для создания массивов определённой формы:

np.zeros(n) — массив из нулей

np.ones(n) — массив из единиц

np.arange(start, stop, step) — диапазон чисел

np.linspace(start, stop, num) — равномерно распределённые значения Пример: np.zeros(5) создаст [0, 0, 0, 0, 0]

5.Индексирование массивов NumPy. Индекс и срез Элементы массива можно получать по индексу (начиная с 0). Срез позволяет брать часть массива: массив[начало:конец:шаг]. Пример: a = np.array([10,20,30,40,50]); a[1:4] → [20,30,40]

6.Операции с массивами. Универсальные (поэлементные) функции NumPy позволяет выполнять арифметические операции сразу со всем массивом, без циклов. Примеры: a + b, a * 2, np.sqrt(a), np.exp(a), np.sin(a) Такие функции применяются к каждому элементу массива одновременно.



7.Функция mean() Вычисляет среднее арифметическое элементов массива. Синтаксис: np.mean(массив) Пример: np.mean([1,2,3]) → 2.0

8.Функция median() Находит медиану — центральное значение отсортированного массива. Синтаксис: np.median(массив) Пример: np.median([1,3,2]) → 2.0

9.Функция corrcoef() Вычисляет коэффициент корреляции между массивами, показывая степень связи данных. Пример:

np.corrcoef([1,2,3],[2,4,6])

Результат покажет сильную положительную корреляцию (значение близкое к 1).

10.Дисперсия var() Показывает, насколько элементы массива отклоняются от среднего значения. Синтаксис: np.var(массив) Пример: np.var([1,2,3,4]) → 1.25

11.Стандартное отклонение std() Показывает среднюю величину отклонений элементов от среднего. Синтаксис: np.std(массив) Пример: np.std([1,2,3,4]) → 1.118
